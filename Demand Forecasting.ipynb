### Feature Extraction for working upon on Dataframe

```python
features=['order date (DateOrders)','Sales','Order Item Discount','Order Item Quantity']
```

```python
df=df[features]
```

```python
df.head(5)
```

### Applying to Pandas DateTime format for easy access

```python
df['order date (DateOrders)'] =  df['order date (DateOrders)'].apply(pd.to_datetime)
```

```python
df.head(5)
```

### Sorting the records by date and their time

```python
df = df.sort_values(by='order date (DateOrders)')
```

```python
df.head(10)
```

### Making column of Order date to Index

```python
df.set_axis(df['order date (DateOrders)'], inplace=True)
```

```python
df.drop(columns=['order date (DateOrders)'],inplace=True)
```

```python
df.head(10)
```

```python
df.tail()
```

### Month-wise Aggregates

```python
monthlysales = df.groupby([(df.index.year), (df.index.month)])['Sales'].sum().astype(int)
```

```python
monthlydiscount = df.groupby([(df.index.year), (df.index.month)])['Order Item Discount'].sum().astype(int)
```

```python
monthlyquantity = df.groupby([(df.index.year), (df.index.month)])['Order Item Quantity'].sum().astype(int)
```

### Monthly Sales

```python
print('Monthly Sales across the Years:')
print(monthlysales)
```

### Monthly Discount

```python
print('Monthly Discount across the Years:')
print(monthlydiscount)
```

### Monthly Quantity

```python
print('Monthly Quantity across the Years:')
print(monthlyquantity)
```

### Display in single Dataframe 

```python
totalsales=pd.DataFrame(monthlysales)
totaldiscount=pd.DataFrame(monthlydiscount)
totalquantitiy=pd.DataFrame(monthlyquantity)
```

```python
total=totalsales.join(totaldiscount)
```

```python
total=total.join(totalquantitiy)
```
